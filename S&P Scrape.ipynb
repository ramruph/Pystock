# Scraping the S&P 500 ticker symbols and data

```python
import bs4 as bs
import pickle #used to save s&p 500 list
import requests
```

```python
def save_sp500_tickers():
    
    #get the source code from website
    response = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
    
    #BeautifulSoup object
    soup = bs.BeautifulSoup(response.text)
    
    
    table = soup.find('table', {'id': 'constituents'})
    
    tickers =[]
    
    for row in table.findAll('tr')[1:]:
        
        ticker = row.find('td').text
        
        tickers.append(ticker)
        
    with open("sp500ticker.pickle", "wb") as f:
        pickle.dump(tickers, f)
        
    print (tickers)
    
    return tickers
```

```python
save_sp500_tickers()
```
