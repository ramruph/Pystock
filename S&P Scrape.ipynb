# Scraping the S&P 500 ticker symbols and data

```python
import bs4 as bs
import datetime as dt
import pickle #used to save s&p 500 list
import requests
import pandas as pd
import pandas_datareader.data as web
import os
```

```python
def save_sp500_tickers():
    
    #get the source code from website
    response = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
    
    #BeautifulSoup object
    soup = bs.BeautifulSoup(response.text)
    
    
    table = soup.find('table', {'id': 'constituents'})
    
    tickers =[]
    
    for row in table.findAll('tr')[1:]:
        
        ticker = row.find('td').text
        
        tickers.append(ticker)
        
    with open("sp500ticker.pickle", "wb") as f:
        pickle.dump(tickers, f)
        
    print (tickers)
    
    return tickers
```

```python
#create function to get Yahoo data

def get_data_from_yahoo(reload_sp500=False):
    
    if reload_sp500:
        ticker= save_sp500_tickers()   
    else:
        with open("sp500ticker.pickle", "rb") as f:
            tickers = pickle.load(f)
    
    if not os.path.exists('stock_dfs'):
        os.makedirs('stock_dfs')
    
    #Start and end time
    start = dt.datetime(2000,1,1)
    end = dt.datetime(2020,12,1)
    
    for ticker in tickers:
        if not os.path.exists('stock_dfs/{}.csv'.format(ticker)):
        
            try:
                df = web.DataReader(ticker.strip('\n'),'yahoo', start, end)
            
                df.to_csv('stock_dfs/{}.csv'.format(ticker))
            
            except KeyError:
                pass
        
        else:
            print("Already have {}".format(ticker ))
```
